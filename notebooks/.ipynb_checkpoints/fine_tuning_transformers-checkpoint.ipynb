```python
training_args = TFTrainingArguments(
    output_dir='../models/bertweet_finetuning/results',         
    num_train_epochs=3,             
    per_device_train_batch_size=512,  
    per_device_eval_batch_size=512,   
    warmup_steps=500,                
    weight_decay=0.01,
    learning_rate=1e-05,
    logging_dir='../models/bertweet_finetuning/logs',            
    logging_steps=10,
)

with training_args.strategy.scope():
    model = TFAutoModelForSequenceClassification.from_pretrained("vinai/bertweet-base", num_labels=1)

trainer = TFTrainer(
    model=model,                         
    args=training_args,                  
    train_dataset=train_dataset,         
    eval_dataset=val_dataset             
)
```
